```python
############  ONLINE SHOPPING BEHAVIOUR ANALYSIS

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


shopping_df = pd.DataFrame({
    "Product_Purchases": ["Shoes", "Earbuds", "Accessories", "Mobiles", "Earbuds"],
    "Cart_Abondement": [100, 50, 1, 200, 75],
    "Customer_Ratings": [2, 1, 5, 4, 3],
    "Payment_Methods": ["Online", "Offline", "Online", "Online", "Offline"]
})

print(shopping_df)

##### Most Sales Product
most_sales_product = shopping_df.loc[shopping_df['Cart_Abondement'].idxmax(), 'Product_Purchases']
print("\nMost Sales Product:", most_sales_product)

###### HIGH REVENUE CUSTOMERS
high_revenue_customers = shopping_df.groupby("Payment_Methods")["Cart_Abondement"].sum().idxmax()
print("\nHigh Revenue Customers:", high_revenue_customers)

##### MOST POPULAR PAYMENT METHOD
most_popular_payment_method = shopping_df["Payment_Methods"].mode().values[0]
print("\nMost Popular Payment Method:", most_popular_payment_method)

###### PRODUCTS WITH POOR RATING
poor_rating_products = shopping_df[shopping_df["Customer_Ratings"] < 3]["Product_Purchases"].tolist()
print("\nProducts with Poor Rating:", poor_rating_products)
```

```python
plt.bar(shopping_df['Product_Purchases'], shopping_df['Cart_Abondement'])
plt.xlabel('Product Purchases')
plt.ylabel('Cart Abondement')
plt.title('Cart Abondement by Product')
plt.show()
```

**Bank Transaction Analysis**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame({"Transaction_ID": [1, 2, 3, 4, 5],
                            "Transaction_Amount":[8800,5040,2024,3840,9450],
                            "Location":["Banglore","Hyd","Mumbai","Tamilnadu","Madanapalli"],
                            "Time":["11:30","2:40","10:20","7:00","6:00"],
                            "Acc_Activity":["Actice","Inactive","Active","Blocked","Blocked"]})
print(df)
### Unusual transactions
unusual_transactions = df.loc[
    df['Transaction_Amount'] == df['Transaction_Amount'].max(),
    'Transaction_ID'
].tolist()
print("Unusual Transactions:", unusual_transactions)
##### HIGH RISK ACCOUNTS
high_risk_accounts = df[df['Acc_Activity'] == 'Blocked']['Transaction_ID'].tolist()
print("High Risk Accounts:", high_risk_accounts)
#####FRAUD PATTERNS
fraud_patterns = df[
    (df['Transaction_Amount'] > df['Transaction_Amount'].mean()) &
    (df['Acc_Activity'] == 'Blocked')
]['Transaction_ID'].tolist()
print("Fraud Patterns:", fraud_patterns)
```

```python
plt.bar(df['Location'], df['Transaction_Amount'])
plt.xlabel('Location')
plt.ylabel('Transaction Amount')
plt.title('Transaction Amount by  Location')
plt.show()
```

**SuperMarket Inventory & Sales Analysis**

```python
#DAILY SALE
#PRODUCT STOCK
#EXPIRY DATES
#SEASONAL DEMANDS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.DataFrame({
    "Daily_Sales": [120, 80, 150, 200, 60],
    "Product_Name": ["Milk", "Bread", "Eggs", "Soft Drinks", "Yogurt"],
    "Product_Stock": [50, 30, 100, 150, 20],
    "Expiry_Dates": ["2026-06-05", "2026-06-02", "2026-06-10", "2027-01-15", "2026-06-03"],
    "Seasonal_Demands": ["Medium", "Low", "High", "High", "Medium"]
})

print(df)

#####  Need to identify  Fast Moving Products

fast_moving_products = df.loc[df['Daily_Sales'].idxmax(), 'Product_Name']
print("\nFast Moving Products:", fast_moving_products)
#####    Slow Moving Products
slow_moving_products = df.loc[df['Daily_Sales'].idxmin(), 'Product_Name']
print("\nSlow Moving Products:", slow_moving_products)
###### Outofstock Trends
out_of_stock = df[df['Product_Stock'] < df['Daily_Sales']]['Product_Name'].tolist()
high_seasonal_demand = df[df['Seasonal_Demands'] == 'High']['Product_Name'].tolist()
print("\nHigh Seasonal Demand Products:", ', '.join(high_seasonal_demand))
print("\nOut of Stock Trends:", ', '.join(out_of_stock))
```

```python
plt.bar(df['Product_Name'], df['Daily_Sales'])
plt.xlabel('Product Name')
plt.ylabel('Daily Sales')
plt.title('Daily Sales by Product')
plt.show()
#
```
