# WEATHER PREDICTION DATA

```python
plt.plot(data['weather'],data['precipitation'])
plt.xlabel("weather")
plt.ylabel("precipitation")
plt.title("line chart")
plt.show()
```

```python
plt.bar(data['weather'],data['temp_max'])
plt.xlabel("weather")
plt.ylabel("temp_max")
plt.title("bar chart")
plt.show()
```

```python
plt.scatter(data['temp_max'],data['temp_min'])
plt.xlabel("temp_max")
plt.ylabel("temp_min")
plt.title("scatter chart")
plt.show()
```
