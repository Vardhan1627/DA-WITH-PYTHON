TYPES OF DATA ANALYTICS WITH PYTHON


**DESCRIPTIVE ANALYSIS**

```python
import statistics
import matplotlib.pyplot as plt
names=["AKBAR","ANTONY","SHANMUKH","KAMAL","SHIVAJI"]
ages=[22,32,29,21,22]
wrk_hours=[2,4,3.5,6,2]
calories_burn=[220,280,264,500,750]
## FOR DISPLAYING DATASET
print("\n----------FITNESS CENTER DATASET-----------\n")
for i in range(len(names)):
  print(names[i],
        ages[i],
        wrk_hours[i],
        calories_burn[i]
        )

## MEAN WORKOUT HOURS
avg_workout=sum(wrk_hours)/len(wrk_hours)
print("\nAverage Workout Hours=", avg_workout)

### MAX CALORIES
high_calories=max(calories_burn)
print("Maximum Calories Burned=", high_calories)

### MIN CALORIES
low_calories=min(calories_burn)
print("Minimum Calories Burned=", low_calories)

## STANDARD DEVIATION
std_dev=statistics.stdev(calories_burn)
print("Standard Deviation=", std_dev)

## MOST ACTIVE PERSON
index=calories_burn.index(max(calories_burn))
print("\nMost Active Person=", names[index])

## LEAST ACTIVE PERSON
index=calories_burn.index(min(calories_burn))
print("Least Active Person=", names[index])
```

```python
## BARCHART
plt.bar(names,calories_burn)
plt.xlabel("Names")
plt.ylabel("Calories Burned")
plt.title("FITNESS CENTER DATASET")
plt.show()
```

```python
### PIECHART
plt.pie(wrk_hours,labels=names)
plt.title("FITNESS CENTER DATASET")
plt.show()
```

```python
plt.plot(names, wrk_hours)
plt.title("FITNESS CENTER DATASET")
plt.show()
```

```python
plt.boxplot([[c] for c in calories_burn], tick_labels=names)
plt.title("FITNESS CENTER DATASET")
plt.show()
```
