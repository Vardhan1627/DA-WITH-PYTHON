**By Using Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/heart.csv", engine='python')

# Show columns
print(df.columns)

# Select first numeric column
col = df.select_dtypes(include='number').columns[0]

print("Using Column:", col)
# Average
avg = df[col].mean()
print("Average Value:", avg)

# Prescriptive Output
for i in df[col].head(10):

    if i > avg:
        print(i, "-> High Value")

    else:
        print(i, "-> Normal Value")
```

```python
plt.plot(df[col])
plt.title("Line Chart")
plt.xlabel("Index")
plt.ylabel(col)
plt.show()
```

```python
high = len(df[df[col] > avg])
low = len(df[df[col] <= avg])

plt.pie([high, low],
        labels=["High", "Normal"],
        autopct='%1.1f%%')

plt.title("High vs Normal")

plt.show()
```

```python
df[col].head(10).plot(kind='bar')

plt.title("Bar Chart")
plt.xlabel("Index")
plt.ylabel(col)
```

```python
sns.histplot(df[col])

plt.title("Histogram")

plt.show()
```

**With User Dataset performing**

```python
import pandas as pd

data = {
    "PatientName": ["Shanmukha", "Arun", "Bhavan", "Suriya"],
    "Age": [65, 25, 70, 40],
    "OxygenLevel": [72, 95, 80, 88]
}

df = pd.DataFrame(data)

score = []

for i in range(len(df)):

    s = 0

    if df.loc[i, "Age"] > 60:
        s += 2

    if df.loc[i, "OxygenLevel"] < 85:
        s += 3

    score.append(s)

df["PriorityScore"] = score

bed = []

for i in df["PriorityScore"]:

    if i >= 3:
        bed.append("Allocated")

    else:
        bed.append("Not Allocated")

df["Bed"] = bed

prescription = []

for i in df["PriorityScore"]:

    if i >= 5:
        prescription.append("ICU")

    elif i >= 3:
        prescription.append("Oxygen Support")

    else:
        prescription.append("Normal Checkup")

df["Prescription"] = prescription

print(df)
```

```python
plt.plot(df["OxygenLevel"])
plt.title("Line Chart")
plt.xlabel("Condition")
plt.ylabel("OxygenLevel")
plt.show()
```

```python
df["PriorityScore"] = df["PriorityScore"].replace(0,1)

plt.pie(
    df["PriorityScore"],
    labels=df["PatientName"],
    autopct='%1.1f%%'
)

plt.show()
```

```python
plt.bar(df["PatientName"], df["OxygenLevel"])

plt.title("Bar Chart")
plt.xlabel("PatientName")
plt.ylabel("OxygenLevel")
```

```python

plt.hist(df["OxygenLevel"])

plt.title("Histogram")

plt.show()
```

```python
plt.scatter(df["Age"], df["PriorityScore"])

plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("PriorityScore")
```

```python
plt.boxplot(df["OxygenLevel"])

plt.title("Box Plot")
plt.ylabel("OxygenLevel")
```
