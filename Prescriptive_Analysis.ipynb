**By Using Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/AguaH.csv", engine='python')

# Show columns
print(df.columns)

# Select first numeric column
col = df.select_dtypes(include='number').columns[0]

print("Using Column:", col)

# Average
avg = df[col].mean()

print("Average Value:", avg)

# Prescriptive Output
for i in df[col].head(10):

    if i > avg:
        print(i, "-> High Value")

    else:
        print(i, "-> Normal Value")
```

```python
plt.plot(df[col])
plt.title("Line Chart")
plt.xlabel("Index")
plt.ylabel(col)
plt.show()
```

```python
high = len(df[df[col] > avg])
low = len(df[df[col] <= avg])

plt.pie([high, low],
        labels=["High", "Normal"],
        autopct='%1.1f%%')

plt.title("High vs Normal")

plt.show()
```

```python
df[col].head(10).plot(kind='bar')

plt.title("Bar Chart")
plt.xlabel("Index")
plt.ylabel(col)
```

```python
sns.histplot(df[col])

plt.title("Histogram")

plt.show()
```

**With User Dataset performing**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating sample values manually
data = {
    "Water_Consumption": [45, 60, 120, 80, 150, 30, 95, 110, 70, 140]
}

# Create DataFrame
df = pd.DataFrame(data)

print(df)

# Average
avg = df["Water_Consumption"].mean()

print("\nAverage Water Consumption:", avg)

# Prescriptive Analysis
for i in df["Water_Consumption"]:

    if i > 120:
        print(i, "-> Critical Usage : Reduce Water Usage")

    elif i > avg:
        print(i, "-> High Usage : Monitor Usage")

    else:
        print(i, "-> Normal Usage : Maintain Supply")
```

```python
plt.plot(df["Water_Consumption"])
plt.title("Water Consumption Trend")
plt.show()
```

```python
sns.histplot(df["Water_Consumption"])
plt.title("Distribution")
plt.show()
```

```python
sns.boxplot(x=df["Water_Consumption"])
plt.title("Boxplot")
plt.show()
```

```python

plt.scatter(df.index, df["Water_Consumption"])
plt.title("Scatter Plot")
plt.show()
```
