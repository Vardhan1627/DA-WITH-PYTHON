TYPES OF DATA ANALYTICS WITH PYTHON


**DESCRIPTIVE ANALYSIS**

```python
import statistics
import matplotlib.pyplot as plt
names=["AKBAR","ANTONY","SHANMUKH","KAMAL","SHIVAJI"]
ages=[22,32,29,21,22]
wrk_hours=[2,4,3.5,6,2]
calories_burn=[220,280,264,500,750]
## FOR DISPLAYING DATASET
print("\n----------FITNESS CENTER DATASET-----------\n")
for i in range(len(names)):
  print(names[i],
        ages[i],
        wrk_hours[i],
        calories_burn[i]
        )

## MEAN WORKOUT HOURS
avg_workout=sum(wrk_hours)/len(wrk_hours)
print("\nAverage Workout Hours=", avg_workout)

### MAX CALORIES
high_calories=max(calories_burn)
print("Maximum Calories Burned=", high_calories)

### MIN CALORIES
low_calories=min(calories_burn)
print("Minimum Calories Burned=", low_calories)

## STANDARD DEVIATION
std_dev=statistics.stdev(calories_burn)
print("Standard Deviation=", std_dev)

## MOST ACTIVE PERSON
index=calories_burn.index(max(calories_burn))
print("\nMost Active Person=", names[index])

## LEAST ACTIVE PERSON
index=calories_burn.index(min(calories_burn))
print("Least Active Person=", names[index])

## BARCHART
plt.bar(names,calories_burn)
plt.xlabel("Names")
plt.ylabel("Calories Burned")
plt.title("FITNESS CENTER DATASET")
plt.show()

### PIECHART
plt.pie(wrk_hours,labels=names)
plt.title("FITNESS CENTER DATASET")
plt.show()

## LINECHART
plt.plot(names, wrk_hours)
plt.title("FITNESS CENTER DATASET")
plt.show()

### BOXPLOT
plt.boxplot([[c] for c in calories_burn], tick_labels=names)
plt.title("FITNESS CENTER DATASET")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
# data = pd.read_csv("/content/StudentsPerformance.csv") # This file was not found
# For demonstration purposes, creating a dummy DataFrame if the file isn't loaded
data = pd.DataFrame({
    'gender': ['male', 'female', 'male', 'female'],
    'race/ethnicity': ['group A', 'group B', 'group C', 'group A'],
    'math score': [70, 65, 80, 75],
    'reading score': [75, 70, 85, 80],
    'writing score': [72, 68, 82, 78]
})
print(data)
data.drop_duplicates(inplace=True)
# missing values
print("MISSING VALUES")
print(data.isnull().sum())
# Remove duplicates
print("REMOVING DUPLICATES")
data.drop_duplicates(inplace=True)
# Cleaning data
data.to_csv("cleaned_data.csv", index=False)

print("Data cleaned successfully")
```

```python
plt.pie(data["gender"].value_counts(), labels=["Male", "Female"], autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
plt.bar(data["race/ethnicity"], data["math score"])
plt.xlabel("Race/Ethnicity")
plt.ylabel("Math Score")
plt.title("Race/Ethnicity vs Math Score")
plt.show()
```

```python
plt.scatter(data["writing score"], data["reading score"])
plt.xlabel("writing score")
plt.ylabel("Reading Score")
plt.title("Writing Score vs Reading Score")
plt.show()
```

```python
plt.hist(data["math score"], bins=10)
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.title("Distribution of Math Scores")
plt.show()
```
