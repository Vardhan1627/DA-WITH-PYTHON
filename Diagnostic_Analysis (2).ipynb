# DIAGNOSTIC ANALYSIS from User data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df=pd.DataFrame({
"names":["Vardhan","Shivaram","Sunil","Shanmukha","Vishnu"],
"products":["Mobile","Laptop","Dress","Shoes","Facewash"],
"price":[19000,60000,1000,500,2000],
"sales":[200,100,300,50,150],
"speed":[123,150,200,155,160],
"pay_fail":[0,1,0,1,0],
"Del_charges":[100,50,80,150,100],
"pur_completed":[True,False,True,False,True],
})
print(df)
total_abandnment=df['pur_completed'].value_counts()[False]
print("total_abandnment:",total_abandnment)


internet_diagnosis=(df['speed'] < 150).sum()
print("internet_diagnosis (low speed):",internet_diagnosis)

payment_diagnosis=df['pay_fail'].value_counts()[1]
print("payment_diagnosis (failure):",payment_diagnosis)

delivery_diagnosis=(df['Del_charges'] > 50).sum()
print("delivery_diagnosis (charges > 50):",delivery_diagnosis)

print("ROOT CAUSES ARE:")
for i in range(len(df['names'])):
  if df['speed'][i] < 150:
    print("low internet speed")
  elif df['pay_fail'][i]==1:
    print("payment failed")
  elif df['Del_charges'][i]>70:
    print("high delivery charges")
```

```python
plt.plot(df["products"],df["sales"])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Products vs Sales")
plt.show()
```

```python
plt.bar(df["products"],df["sales"])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Products vs Sales")
plt.show()
```

```python
plt.scatter(df["Del_charges"],df["pur_completed"])
plt.xlabel("Del_charges")
plt.ylabel("pur_completed")
plt.title("Del_charges vs pur_completed")
plt.show()
```

```python
plt.hist(df["sales"])
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
plt.pie(df["sales"],labels=df["products"],autopct="%1.2f%%")
plt.legend(loc="best")
plt.title("Pie Chart")
plt.show()
```

```python
plt.boxplot(df["sales"])
plt.title("Boxplot")
plt.show()
```

# Diagnostic Analysis from Kaggle Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv("/content/sales_data.csv")
print(data)
```

```python
total_abandnment=data['Payment_Method'].value_counts()['Cash']
print("total_abandnment:",total_abandnment)

internet_diagnosis=(data['Unit_Cost'] < 150).sum()
print("internet_diagnosis:",internet_diagnosis)

payment_diagnosis=data['Quantity_Sold'].value_counts()[1]
print("payment_diagnosis:",payment_diagnosis)

delivery_diagnosis=(data['Sales_Amount'] > 5000).sum()
print("delivery_diagnosis (Sales_Amount > 5000):",delivery_diagnosis)

print("ROOT CAUSES ARE:")
for i in range(5):
  if data['Quantity_Sold'][i] < 30:
    print("Low Quantity sold")
  elif data['Discount'][i]==0:
    print("No discount")
  elif data['Sales_Amount'][i]>5000:
    print("high sales amount")
```

```python
plt.plot(data["Region"],data["Quantity_Sold"])
plt.xlabel("Region")
plt.ylabel("Quantity_Sold")
plt.title("Region vs Quantity_Sold")
plt.show()
```

```python
plt.scatter(data["Sales_Amount"],data["Discount"])
plt.xlabel("Sales_Amount")
plt.ylabel("Discount")
plt.title("Sales_Amount vs Discount")
plt.show()
```

```python
plt.bar(data["Product_Category"],data["Quantity_Sold"])
plt.xlabel("Product_Category")
plt.ylabel("Quantity_Sold")
plt.title("Product_Category vs Quantity_Sold")
plt.show()
```

```python
plt.hist(data["Quantity_Sold"])
plt.xlabel("Quantity_Sold")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
payment_method_counts = data["Payment_Method"].value_counts()
plt.pie(payment_method_counts.values, labels=payment_method_counts.index, autopct="%1.1f%%")
plt.legend(loc="best")
plt.title("Distribution of Payment Methods")
plt.show()
```

```python
plt.boxplot(data["Quantity_Sold"])
plt.title("Boxplot")
plt.show()
```
