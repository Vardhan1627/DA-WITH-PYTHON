```python
import pandas as od
import numpy as np
marks=[70,80,90,60,100]
hours=[2,3,5,1,6]
print("Mean:",np.mean(marks))
print("Median:",np.median(marks))
print("Standard deviation:",np.std(marks))
print("Correlation coefficient:",np.corrcoef(marks,hours))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("/content/Sales Dataset.csv")
print(df)
```

```python
plt.hist(df["Profit"],color="red")
plt.xlabel("profit")
plt.ylabel("frequency")
plt.title("Histogram")
plt.show()
```

```python
plt.bar(df["Profit"],df["City"])
plt.xlabel("profit")
plt.ylabel("City")
plt.title("bar chart")
plt.show()
```

```python
plt.boxplot(df["Profit"])
plt.xlabel("profit")
plt.ylabel("frequency")
plt.title("box plot")
plt.show()
```

```python
plt.pie(df["Profit"])
plt.xlabel("profit")
plt.ylabel("City")
plt.title("pie chart")
plt.show()
```

```python
plt.scatter(df["Profit"],df["City"])
plt.xlabel("profit")
plt.ylabel("City")
plt.title("scatter plot")
plt.show()
```
