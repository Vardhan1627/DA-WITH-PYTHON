```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame({"Water_Consumption":[143,234,454,234,123],
                 "Days":["Monday","Tuesday","Wednesday","Thrsday","Friday",]})
print(df)
avg_of_water=df["Water_Consumption"].mean()
print("Avg_Water:",avg_of_water)
total_water_usage=df["Water_Consumption"].sum()
print("Water_Usage:",total_water_usage)
# Calculate the number of days as an integer count, as was done previously.
sum_of_days=len(df)
print("No_Of_Days:",sum_of_days)
avg_water_usage=total_water_usage/sum_of_days
print("Avg_Water_Usage:",avg_water_usage)
shortage_water=avg_of_water-avg_water_usage
print("Shortage_water:",shortage_water)
```

```python
plt.bar(df["Days"],df["Water_Consumption"])
plt.xlabel("Days")
plt.ylabel("Water Consumption")
plt.title("Water Consumption by Day")
plt.show()
```

```python
plt.pie(df["Water_Consumption"],labels=df["Days"],autopct="%1.1f%%")
plt.title("Water Consumption by Day")
plt.show()
```

```python
plt.hist(df["Water_Consumption"])
plt.xlabel("Water Consumption")
plt.ylabel("Frequency")
plt.title("Water Consumption Histogram")
plt.show()
```

```python
plt.boxplot(df["Water_Consumption"])
plt.xlabel("Water Consumption")
plt.title("Water Consumption Box Plot")
plt.show()
```
